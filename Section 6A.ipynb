<p>
Define the probability density function (PDF) for the nearest star distance.
</p>

<p>
For a uniform Poisson process in 3D, the probability that no star is found within a sphere of radius \( R \) is:
</p>

$$
P(\text{no stars in sphere}) = \exp\left(-\frac{4}{3}\pi n R^3\right)
$$

<p>
Therefore, the cumulative probability that the nearest star is within distance \( R \) is:
</p>

$$
\text{CDF}(R) = 1 - \exp\left(-\frac{4}{3}\pi n R^3\right)
$$

<p>
Differentiating the CDF with respect to \( R \) gives the PDF:
</p>

$$
\text{PDF}(R) = \frac{d}{dR}\left[1 - \exp\left(-\frac{4}{3}\pi n R^3\right)\right] = 4\pi n R^2 \exp\left(-\frac{4}{3}\pi n R^3\right)
$$

```python
n = 0.01  # You can adjust this value as needed


def nearest_star_pdf(R, n):
    return 4 * np.pi * n * R**2 * np.exp(- (4/3) * np.pi * n * R**3)

R_values = np.linspace(0, 50, 500)  # Adjust the range as needed
pdf_values = nearest_star_pdf(R_values, n)
plt.figure(figsize=(8, 6))
plt.plot(R_values, pdf_values, label=r'$P(R) = 4 \pi n R^2 e^{-\frac{4}{3}\pi n R^3}$')
plt.xlabel('Distance R')
plt.ylabel('Probability Density')
plt.title('Probability Density Function for the Nearest Star Distance')
plt.legend()
plt.grid(True)
plt.show()
```

<p>
Consider the driven, damped harmonic oscillator:
</p>

$$
\frac{d^2x}{dt^2} + \gamma \frac{dx}{dt} + \omega_0^2 x = F e^{i\omega_f t}
$$

<p>
Taking the Fourier transform yields:
</p>

$$
(-\omega_f^2 + i\gamma\omega_f + \omega_0^2)\tilde{x} = F
$$

<p>
Thus, the Fourier amplitude is:
</p>

$$
\tilde{x} = \frac{F}{\omega_0^2 - \omega_f^2 + i\gamma\omega_f}
$$

<p>
Defining the susceptibility as
</p>

$$
\chi(\omega_f) = \frac{1}{\omega_0^2 - \omega_f^2 + i\gamma\omega_f},
$$

<p>
its imaginary part is:
</p>

$$
\operatorname{Im}[\chi(\omega_f)] = \frac{\gamma \omega_f}{(\omega_0^2 - \omega_f^2)^2 + \gamma^2 \omega_f^2}
$$

<p>
Therefore, the energy absorbed per cycle is given by:
</p>

$$
E = F\pi\, \operatorname{Im}[\chi(\omega_f)] = \frac{F\pi\gamma\omega_f}{(\omega_0^2 - \omega_f^2)^2 + \gamma^2\omega_f^2}
$$

<b>Question 4: Markov Chain in Magnon Basis</b>
</p>

<p>
For a chain of \(N=3\) spins, the allowed magnon momenta are quantized as 

$$
p_k = \frac{2\pi k}{3},\quad k=0,1,2,
$$
with corresponding energies 

$$
E_k = 2J\sin^2\left(\frac{\pi k}{3}\right).
$$
Thus,

$$
E_0 = 0,\quad E_1 = 2J\sin^2\left(\frac{\pi}{3}\right) = \frac{3J}{2},\quad E_2 = \frac{3J}{2}.
$$
</p>

<p>
Assuming Boltzmann-type transitions between magnon states, we take the transition probability from state \(i\) to state \(j\) as

$$
P_{ij} \sim e^{-\frac{E_j - E_i}{k_BT}}.
$$
Normalizing each row, we define:
</p>

<p>
For state \(0\):
</p>

$$
P_{0j} = \frac{e^{-\frac{E_j-0}{k_BT}}}{Z_0},\quad Z_0 = e^0 + 2e^{-\frac{3J}{2k_BT}} = 1+2e^{-\frac{3J}{2k_BT}}.
$$
<p>
For state \(1\) (and similarly state \(2\)):
</p>

$$
P_{1j} = \frac{e^{-\frac{E_j-E_1}{k_BT}}}{Z_1},\quad Z_1 = e^{\frac{3J}{2k_BT}} + 1 + 1 = e^{\frac{3J}{2k_BT}}+2.
$$
<p>
Thus, the transition matrix in the magnon basis is:
</p>

$$
P = \begin{pmatrix}
\frac{1}{1+2e^{-\frac{3J}{2k_BT}}} & \frac{e^{-\frac{3J}{2k_BT}}}{1+2e^{-\frac{3J}{2k_BT}}} & \frac{e^{-\frac{3J}{2k_BT}}}{1+2e^{-\frac{3J}{2k_BT}}} \\[1em]
\frac{e^{\frac{3J}{2k_BT}}}{e^{\frac{3J}{2k_BT}}+2} & \frac{1}{e^{\frac{3J}{2k_BT}}+2} & \frac{1}{e^{\frac{3J}{2k_BT}}+2} \\[1em]
\frac{e^{\frac{3J}{2k_BT}}}{e^{\frac{3J}{2k_BT}}+2} & \frac{1}{e^{\frac{3J}{2k_BT}}+2} & \frac{1}{e^{\frac{3J}{2k_BT}}+2}
\end{pmatrix}.
$$
<p>
The key difference relative to the site-basis transition matrix is that here the probabilities are weighted by energy differences (via Boltzmann factors), reflecting the thermal population of the magnon states.
</p>

<b>Question 5: Stationary Distribution in the Magnon Basis</b>
</p>

<p>
The stationary distribution \(\pi\) satisfies

$$
\pi P = \pi,\quad \sum_{k=0}^{2}\pi_k = 1.

Because the system obeys detailed balance, the stationary distribution is given by the Boltzmann distribution:
</p>

$$
\pi_k = \frac{e^{-\frac{E_k}{k_BT}}}{\sum_{k'=0}^{2} e^{-\frac{E_{k'}}{k_BT}}}.

<p>
For \(N=3\) with \(E_0=0\) and \(E_1=E_2=\frac{3J}{2}\), we have:
</p>

$$
\pi_0 = \frac{1}{1+2e^{-\frac{3J}{2k_BT}}},\quad \pi_1 = \pi_2 = \frac{e^{-\frac{3J}{2k_BT}}}{1+2e^{-\frac{3J}{2k_BT}}}.

<p>
Power iteration evolves the distribution by

<b>Question 7: Master Equation Evolution</b>
</p>

<p>
The continuous-time master equation is given by:
</p>

$$
\frac{d\pi_i}{dt} = \sum_{j} \left(Q_{ji}\pi_j - Q_{ij}\pi_i\right).
$$

<p>
To relate the discrete-time transition matrix \(P\) to the continuous-time transition rate matrix \(Q\), we use:
</p>

$$
P \approx e^{Q\Delta t}\quad \Longrightarrow \quad Q \approx \frac{1}{\Delta t}\ln(P).
$$

<p>
Once \(Q\) is determined, the master equation in matrix form reads:
</p>

$$
\frac{d\pi}{dt} = \pi Q,
$$
<p>
with an initial condition such as \(\pi(0) = (0,1,0)\) (corresponding to starting in \(|k=1\rangle\)). This system can be solved numerically using standard ODE integrators (for example, with Python’s <code>scipy.integrate.solve_ivp</code> or <code>odeint</code>).
</p>
